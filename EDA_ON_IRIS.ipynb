```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
df = pd.read_csv('D:\Subjects\Python\Projects\iris.csv')
```

```python
# Display the first 5 rows
print(df.head())
```

```python
# Display basic information about the dataset
print(df.info())
```

```python
# Display statistical summary of the dataset
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Histograms for each feature
df.hist(bins=20, figsize=(12, 8))
plt.suptitle("Histograms of Iris Features")
plt.show()
```

```python
# Pair plot for feature distribution and relationships
sns.pairplot(df, hue='Species', markers=['o', 's', 'D'])
plt.suptitle("Pairplot of Iris Features")
plt.show()
```

```python
# Compute the correlation matrix
corr_matrix = df.iloc[:, :-1].corr()

# Plot the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix of Iris Features")
plt.show()
```

```python
# Box plots for each feature by species
plt.figure(figsize=(12, 8))

# Determine the number of features excluding the target variable
num_features = len(df.columns[:-1])
# Calculate the number of rows and columns for subplots
rows = (num_features + 1) // 2  # +1 to account for odd number of features

for i, column in enumerate(df.columns[:-1]):
    plt.subplot(rows, 2, i + 1)
    sns.boxplot(x='Species', y=column, data=df)  # Ensure 'species' matches your column name exactly
    plt.title(f'Boxplot of {column} by Species')

plt.tight_layout()
plt.show()
```

```python
# Violin plots for each feature by species
num_features = len(df.columns[:-1])

# Determine the number of rows needed for the subplots
rows = (num_features + 1) // 2  # This will calculate the minimum rows needed for a 2-column grid

plt.figure(figsize=(12, 8))

for i, column in enumerate(df.columns[:-1]):
    plt.subplot(rows, 2, i + 1)  # Adjust to have enough rows for all features
    sns.violinplot(x='Species', y=column, data=df)  # Ensure 'Species' matches your column name exactly
    plt.title(f'Violin Plot of {column} by Species')

plt.tight_layout()
plt.show()
```
